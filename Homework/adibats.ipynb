Run the following code to plot isotherms and adiabats

```python
import numpy as np
from matplotlib import pyplot

def p_isotherm(T,V):
    return n * R * T / V

def p_adibat(V,gamma):
    return p_isotherm(T2,15) * (15/V)**gamma # 15 l is the starting volume

R = 0.082 # atm * l / mol / K
n = 1 # mole
T1 = 200 # K
T2 = 500 # K
Cp_mono = 5/2 * R # monoatomic gas 
Cv_mono = 3/2 * R # monoatomic gas
Cp_di   = 7/2 * R # diatomic gas 
Cv_di   = 5/2 * R # diatomic gas
gamma_mono = Cp_mono / Cv_mono
gamma_di   = Cp_di   / Cv_di
V = np.linspace(10,100,num = 120)
pyplot.xticks(np.arange(10, 100, step=10))
pyplot.xlim(left=10)
pyplot.xlim(right=100)
pyplot.ylim(top=2.8)
pyplot.ylim(bottom=0.1)
pyplot.plot(V,p_isotherm(T1,V),'-',color='b',label='Isotherm (200 K)')
pyplot.plot(V,p_isotherm(T2,V),'-',color='r',label='Isotherm (500 K)')
pyplot.plot(V,p_adibat(V,gamma_mono),'-',color='k',label='Adiabat monoatomic gas')
pyplot.plot(V,p_adibat(V,gamma_di),'--',color='k',label='Adiabat diatomic gas')
pyplot.ylabel("p (atm)", style='italic')
pyplot.xlabel("V (l)", style='italic')
pyplot.legend()
pyplot.savefig('adiabat.svg')
```
