Run the code below to plot the liquid-vapor phase boundary for water

```python
%matplotlib widget
import numpy as np
from matplotlib import pyplot

deltaH = 40650 # J/mol
R = 8.314 # J/mol/K
T = np.linspace(298,330,num = 120)
p = np.exp(deltaH/R * (1/298 - 1/T))
pyplot.xticks(np.arange(250, 400, step=25))
pyplot.xlim(left=250)
pyplot.xlim(right=400)
pyplot.plot(T,p,'-',color='k')
pyplot.ylabel("p (atm)")
pyplot.xlabel("T (K)")
```
