Numerical solution to Problem 2b of Unit 5

```python
import numpy as np

#Solution of 
M = 78.11e-3 # Kg/mol
deltaV  = 1.3e-5 * M # m^3 / mol
deltaHf = 9.951e3 # J / mol
deltaHv = 30.72e3 # J / mol
Tf_star = 278.64  # K
Tb_star = 353.24  # K
R       = 8.31446261815324 # SI units

#Initial guess
p_triple = 0
for i in range(1,10):
    T_triple = Tf_star / np.exp ( (1 - p_triple) * deltaV / deltaHf )
    p_triple = np.exp( deltaHv / R * (1 / Tb_star - 1 / T_triple) )
    print(i,T_triple,p_triple)
```
