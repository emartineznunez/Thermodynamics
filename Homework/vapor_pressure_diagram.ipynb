Run the code below to plot the mole fraction of A in the vapor $y_{A}$ as a function of the mole fraction of A in the liquid $x_{A}$

```python
import numpy as np
from matplotlib import pyplot

pB = 1 # a.u.
ratio = float(input("pA*/pB* ratio= "))
pA = ratio * pB

xA = np.linspace(0,1,num = 120)
yA = xA * pA / (pB + (pA - pB) * xA)

%matplotlib widget
pyplot.plot(xA,yA,'-',color='k')
pyplot.ylabel("Mole fraction of A in the vapor (yA)")
pyplot.xlabel("Mole fraction of A in the liquid (xA)")
pyplot.xlim(0,1)
```

Run the code below to plot the total vapor pressue $p$ as a function of the mole fraction of A.

Red plot shows $p$ as a function of $x_{A}$ and blue as a function of $y_{A}$

```python
import numpy as np
from matplotlib import pyplot

pB = 1  # a.u.
ratio = float(input("pA*/pB* ratio= "))
pA = ratio * pB

yA = np.linspace(0,1,num = 120)
pv = pA * pB / (pA + (pB - pA) * yA)
pl = pB + (pA - pB) * yA

%matplotlib widget
pyplot.plot(yA,pv,'-',color='b')
pyplot.plot(yA,pl,'-',color='r')
pyplot.ylabel("Total vapor pressure (p in a.u.)")
pyplot.xlabel("Mole fraction of A")
pyplot.savefig("vapor_pressure_diagram.png")
```
