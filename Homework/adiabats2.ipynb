Let's calculate the drop in temperature in a reversible adiabatic expansion of an ideal gas vs the corresponding irreversible one

```python
import numpy as np
from matplotlib import pyplot

def ratio(x):
    return Tfrev / Tfirrev * (x ** gamma - x ) / ( (x - 1) * (gamma - 1) )

V_i = 10.25 # volume in liters
V_f = np.linspace(10.35, 100, 1000)
           
x = V_f / V_i
#gas_type = input("gas type: mono or di: ")

Tfrev = float(input('Final T rev process: '))
Tfirrev = float(input('Final T irrev process: '))


#print(x,ratio(x))
%matplotlib widget
gamma = 1.67
pyplot.plot(x,ratio(x),'-',color='b',label='monoatomic')
gamma = 1.4
pyplot.plot(x,ratio(x),'-',color='r',label='diatomic')
pyplot.legend()
pyplot.ylabel(r'$\Delta T_{rev} / \Delta T_{irrev}$')
pyplot.xlabel(r'$V_f/V_i$')

```
